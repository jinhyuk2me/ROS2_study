```python
def draw_sin(t, A, f, b):
    y = A * np.sin(np.pi * f * t) + b

    plt.figure(figsize=(12,6))
    plt.plot(t,y)
    plt.grid()
    plt.show()
```

```python
t = np.arange(0, 3, 0.01)
draw_sin(t, 1, 1, 0)
```

```python
class DrawSin:
    def __init__(self, amp, freq, bias, end_time):
        self.amp = amp
        self.freq = freq
        self.bias = bias
        self.end_time = end_time
    def calc_sin(self):
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp * np.sin(2*np.pi*self.freq*self.t) + self.bias
    def draw_sin(self):
        y = self.calc_sin()
        plt.figure(figsize=(12,6))
        plt.plot(self.t, y)
        plt.grid()
        plt.show() 
```

```python
ds = DrawSin(1,1,0,3)
ds.draw_sin()
```

```python
ds.freq = 2
ds.draw_sin()
```
